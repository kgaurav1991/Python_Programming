```python
#1. Write a program to Palindrome
num = 'madam'
if num == num[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#2. Write a program to Palindrome
num = 121
sum =0
temp = num

while num != 0:
    r = num % 10
    sum = sum * 10 + r
    num = num // 10
if temp == sum:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Fibonacci Series
num = int(input("Enter the number: "))

a , b = 0 , 1
print("Fibonacci Series: ")
for i in range(num):
    print(a , end = " ")
    c = a + b
    a = b
    b = c

```

```python
# 2. Fibonacci Using Recursion
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)


terms = int(input("Enter number of terms: "))


for i in range(terms):
    print(fibonacci(i), end=" ")
```

```python
# Write a program to prime Number
num = int(input("Enter a number: "))

if num <= 1:
    print("Not a Prime Number")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not a Prime Number")
            break
    else:
        print("Prime Number")
```

Data type Conversion:--

Data type conversion is the conversion of object from one data type to another data type.

implicity:--
An implicity type conversion is a conversion performed by the python interpreter without programmer's intervertion.



```python
a = 4
b = 0.5
c = a + b # without user convertion
print(c)
```

Explicit:--
An Explicit type conversion is User-defined that forces on expression or to be of specife data type.

```python
a = 4
b = "40"
c = a + int(b) # integer User convertion
print(c)
```
